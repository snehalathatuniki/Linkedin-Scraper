```python
# This is the old way which is causing the error:
# y = driver.find_elements_by_class_name('results-context-header__job-count')[0].text

# This is the updated way in accordance with the newer versions of Selenium:
elements = driver.find_elements(By.CLASS_NAME, 'results-context-header__job-count')
if elements:
    y = elements[0].text
    print("Number of job listings:", y)
else:
    print("No job listings found with the specified class.")
```

```python
n=pd.to_numeric(y)
```

```python
#Loop to scroll through all jobs and click on see more jobs button for infinite scrolling

i = 2
while i <= int((n+200)/25)+1: 
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    i = i + 1
    
    try:
#         send=driver.find_element_by_xpath("//button[@aria-label='See more jobs']")
#         driver.execute_script("arguments[0].click();", send)   
#         time.sleep(3)
    
        
        buu=driver.find_elements_by_tag_name("button")
        x=[btn for btn in buu if btn.text=="See more jobs"]
        for btn in x:
                driver.execute_script("arguments[0].click();", btn)
                time.sleep(3)
            
    except:
        pass
        time.sleep(5)
         
```

```python
#Create empty lists for company name and job title

companyname= []
titlename= []
```

```python
from selenium.webdriver.common.by import By

# Create empty lists for company name and job title
companyname = []
titlename = []

# Find company name and append it to the blank list
try:
    # Assuming 'n' is defined somewhere above as the number of elements you expect to find
    companies = driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')
    
    for company in companies:
        companyname.append(company.text)

except IndexError:
    print("no")
```

```python
len(companyname)
```

```python
from selenium.webdriver.common.by import By

# Assuming 'n' is defined somewhere in your code as the number of job titles you expect to find
# Create empty list for job title
titlename = []

# Find title name and append it to the blank list
try:
    titles = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')
    
    for title in titles:
        titlename.append(title.text)

except IndexError:
    print("no")

        
    
    
```

```python
len(titlename)
```

```python
from selenium.webdriver.common.by import By

# Initialize your lists
jobList = []
hrefList = []

# Find job links and append them to a list
jobList = driver.find_elements(By.CLASS_NAME, 'base-card__full-link')

for e in jobList:
    href = e.get_attribute('href')
    if href:  # Making sure href is not None
        hrefList.append(href)

# Now hrefList should contain all the href attributes from the job links
```

```python
hrefList
```

```python
#Create dataframe for company name and title

companyfinal=pd.DataFrame(companyname,columns=["company"])
titlefinal=pd.DataFrame(titlename,columns=["title"])
href = pd.DataFrame(hrefList,columns=["link"])

x=companyfinal.join(titlefinal)
x_final=x.join(href)
x_final
```

```python
#Save file in your directory

x_final.to_csv('linkedin.csv')
```
